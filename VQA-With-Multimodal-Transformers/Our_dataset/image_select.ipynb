```python
import os
import cv2
import numpy as np
import random
import shutil
```

```python

frame_folder = '/content/drive/MyDrive/USC_Course/Multimodal_Probabilistic_Learning_of_Human_CSCI535/Group_Project/video_frames'
output_folder = 'images'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
# select one image randomly from each subfolder
subfolders = [f.path for f in os.scandir(frame_folder) if f.is_dir()]
for subfolder in subfolders:
    images = [f.path for f in os.scandir(subfolder) if f.is_file()]
    if len(images) > 0:
        img = cv2.imread(random.choice(images))
        cv2.imwrite(os.path.join(output_folder, os.path.basename(subfolder) + '.jpg'), img)
        print('saved ' + os.path.join(output_folder, os.path.basename(subfolder) + '.jpg'))
    else:
        print('no images in ' + subfolder)
```
